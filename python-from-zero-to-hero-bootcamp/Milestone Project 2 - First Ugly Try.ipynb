## 如何玩

21點是最受歡迎的賭場遊戲，重點是讓手中的點數總和越接近21點越好，但不能超過21點，且比莊家手中的點數高。遊戲開始時，莊家發給每位玩家兩張牌，發給自己一張牌面向上、一張牌面向下。

## 規則

紙牌依牌面點數計算，K、Q、J以10點計算，A牌可隨玩家決定為1或11點，獲勝賠率為1:1。如果首兩張牌總數為21點(任一10點卡與一隻A牌)，稱為黑傑克 (blackjack)。如果您首兩張牌不是黑傑克，您可以選擇「停牌」(不再叫牌)或指示莊家繼續發牌。將牌押在籌碼下表示停牌，以紙牌摩擦桌面表示叫牌。在亮牌的牌局裡，手掌在賭注後方，掌心向後掃動表示叫牌，左右揮動表示停牌。

不論叫牌或停牌，待所有玩家表態後，莊家會打開他的暗牌並依以下規則決定是否繼續叫牌：莊家總點數在17或17 以上，必須停牌；總點數在16以下，莊家必須叫牌，直到總點數超過17或爆牌 (超過21點)*。如果莊家未爆牌，但您的手牌總數較接近21點，您獲勝；若您的點數比莊家低，莊家獲勝。

*莊家若持有軟17必須繼續叫牌。

如果您和莊家點數相同，即為平手 (push)。莊家會將您的賭注留在下注區。

玩家提示：記住遊戲目標是獲得比莊家更高的點數，但不超過21點。在決定叫牌或停牌之前，注意莊家面朝上的「明牌」。

## 分牌 (Splitting Pairs)

如果您拿到兩張相同點數的牌，您可選擇分成兩手牌，但必須依原先的賭注加一套注，同一局內您可分牌三次 (最多四手牌)。每一手牌內，您可持續叫牌，若叫到A牌，您可選擇再次分牌*。分開的A牌不能算做黑傑克，賠率1:1。

玩家提示：許多玩家認為，如果莊家的明牌不好，分牌是個明智決定。

## 雙倍下注 (Doubling Down)

玩家在拿到首兩張牌後，不論牌型，都可選擇將原先的賭注加至雙倍，但只能再叫一張牌。

玩家提示：當您有10或11點，莊家明牌不好，許多玩家喜歡這種玩法。

## 保險 (Insurance)

如果莊家的明牌是A牌，您可選擇「買保險」，也就是相信莊家的暗牌為10點、莊家握有黑傑克，您可加注相當原先賭注一半的額外賭金。如果莊家的暗牌確實為10點，您的保險賠率為1:2。


## 投降 (Surrender)*

在拿到首兩張牌後，若您覺得您的獲勝機率不高，可在莊家選擇叫牌或停牌以前，決定份投降並繳交一半的賭注。在輪到您選擇叫牌或停牌時，您可口頭告知莊家您已投降。

* 以兩副牌玩21點時，不可再次分開A牌，也不可以投降。

```python
import random

name_map_point_dict = {'Ace':11,'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10}
color_tp = ['Spade','Heart','Diamond','Club']
deck_used = 4
desk_in_order = []
game_status = 1

class Card(object):   
    
    def __init__(self,name,point,color,deckNo):
        self.name = name
        self.point = point
        self.color = color
        self.deckNo = deckNo
        
    def __str__(self):
        print 'Name:{0}, Point:{1}, Color:{2}'.format(self.name,self.point,self.color)
        
class Actor(object):
    
    def __init__(self,name):
    
        self.name = name
        self.card = []
        self.points = 0
        self.wins = 0
        self.losses = 0
        self.busts = 0
        self.status = 'alive'
        
    def hit(self):
        global ful_deck
        card = random.choice(ful_deck)
        self.card.append(card)
        ful_deck.remove(card)
        self.points += card.point
        print 'Hit {0}: {1} {2} from deck no.{3}'.format(self.name, card.color, card.name, card.deckNo)
        self.bustCheck()
        return card,ful_deck
    
    def stand(self):
        self.status = 'stand'
        print '{0} choose to stand.'.format(self.name)
        
    def bustCheck(self):        
        if self.points > 21:
            self.status = 'busted'
            self.losses += 1
            self.busts += 1
            print '{0} just busted.'.format(self.name)        
        
class Player(Actor):
    
    role = 'player'
    
    def __init__(self,name):
        Actor.__init__(self,name)
        self.chips = 100
        self.bits = 0
        print 'Player created.'
        
    def split(self):
        pass
    
    def doubling(self):
        pass
    
    def surrender(self):        
        self.status = 'surrendered'
        print 'Player {0} just surrendered.'.format(self.name) 
        
    def bit(self,bit):
        self.bits = bit

class Dealer(Actor):
    
    role = 'dealer'
    
    def __init__(self,name):
        Actor.__init__(self,name)
        self.chips = 300
        #not counting the first card
        print 'Dealer created.'          

            
def getDecks():
        fulDeck = []
        for no in range(0,deck_used):            
            for color in color_tp:
                for name, point in name_map_point_dict.items():
                    fulDeck.append(Card(name,point,color,no)) 
        return fulDeck
    
ful_deck = getDecks()
    
def round_counter():
    
    global desk_in_order
    
    message = ''
    
    alife_stand_lst = (p for p in desk_in_order if p.status in ['alive','stand'] and p.role == 'player')
    busted_lst = (p for p in desk_in_order if p.status == 'busted' and p.role == 'player')
    surrender_lst = (p for p in desk_in_order if p.status == 'surrendered' and p.role == 'player')
    dealer = desk_in_order[-1]
    
    for player in alife_stand_lst:
        if dealer.status == 'busted':
            player.chips += player.bits
            dealer.chips -= player.bits
            player.wins += 1
            dealer.losses += 1
            message += 'Player {0} win this round!\n'.format(player.name)
        elif player.points < dealer.points <= 21:
            player.chips -= player.bits
            dealer.chips += player.bits  
            player.losses += 1
            dealer.wins += 1
            message += 'Player {0} lose this round!\n'.format(player.name)
        elif player.points > dealer.points:
            player.chips += player.bits
            dealer.chips -= player.bits  
            player.wins += 1
            dealer.losses += 1
            message += 'Player {0} win this round!\n'.format(player.name)
        elif player.points == dealer.points:
            player.chips += player.bits
            message += 'Player {0} draw this round!\n'.format(player.name)
                
    for player in busted_lst:
        player.chips -= player.bits
        dealer.chips += player.bits
        player.losses += 1
        player.busts += 1
        dealer.wins += 1
        message += 'Player {0} lose this round!\n'.format(player.name)
    
    for player in surrender_lst:
        player.chips -= player.bits/2
        dealer.chips += player.bits/2   
        player.losses += 1
        dealer.wins += 1
        message += 'Player {0} lose this round!\n'.format(player.name)
        
    return message
            
def add_player(num):
    global desk_in_order
    while True:
        try:
            name = str(raw_input('Player No.{0}, please advise your name:'.format(num+1)))
        except:
            print 'Your name must include [a-z]'
        else:               
            desk_in_order.append(Player(name))
            break
    return desk_in_order

def ask_bit():
    global desk_in_order
    for p in desk_in_order:
        if p.role == 'player':
            while True:
                try:
                    bit = int(raw_input('Player {0}, please bit:'.format(p.name)))
                except:
                    print 'Please check your bit. Must be integer.'
                else:
                    if bit > p.chips:
                        print 'Your chips is {0}, please bit under this amount.'.format(p.chips)
                        continue
                    else:                        
                        p.bit(bit)
                        break
            
def print_board(is_final):
    global desk_in_order
    for p in desk_in_order:
        if p.role == 'player':
            print '\n{0}: {1}'.format(p.role.upper(),p.name)
            print '----------------'
            print 'Chips: {0}'.format(p.chips)
            print 'Points: {0}'.format(p.points)
            print 'Bits: {0}'.format(p.bits)
            if is_final == True:
                print 'Wins: {0}'.format(p.wins)
                print 'Losses: {0}'.format(p.losses)
                print 'Busts: {0}'.format(p.busts)
            print '----------------'
            for c in p.card:
                print '{0} {1}: {2}pt'.format(c.color,c.name,c.point) 
        elif p.role == 'dealer':
            print '\n{0}: {1}'.format(p.role.upper(),p.name)
            print '----------------'
            print 'Chips: {0}'.format(p.chips)            
            if is_final == True:
                print 'Points: {0}'.format(p.points)
                print 'Wins: {0}'.format(p.wins)
                print 'Losses: {0}'.format(p.losses)
                print 'Busts: {0}'.format(p.busts)
            else:
                print 'Points: {0}'.format(p.points - p.card[0].point)
            print '----------------'
            
            if is_final == True:
                for c in p.card:
                    print '{0} {1}: {2}pt'.format(c.color,c.name,c.point) 
            else:
                print 'Hidden Card...Unkown'
                for c in p.card[1::]:
                    print '{0} {1}: {2}pt'.format(c.color,c.name,c.point) 

def ask_player_for_next():
    global desk_in_order, game_status    
    ps = (p for p in desk_in_order if p.status == 'alive')
    for p in ps:
            if p.role == 'player':            
                split = ''
                hit = 'hit/'
                stand = 'stand/'
                surrender= ''
                if p.card[0].point == p.card[1].point:
                    split = 'split/'
                if game_status == 2:
                    surrender = 'surrender'
                print "\n-----Player {0}'s turn-----".format(p.name)
                while True:
                    try:
                        next = raw_input('{0}, your next move is..?:({1}{2}{3}{4})\n'.format(p.name,split,hit,stand,surrender)).lower()
                    except Exception,e:
                        print str(e)
                    else:
                        if next not in ['split','hit','stand','surrender']:
                            continue
                        else:
                            if next == 'hit':
                                p.hit()
                                break
                            elif next == 'stand':
                                p.stand()
                                break
                            elif next == 'surrender':
                                p.surrender()
                                break

            elif p.role == 'dealer':
                print "\n-----Dealer {0}'s turn-----".format(p.name)
                if p.points <= 17:
                    p.hit()
                else:
                    hit = 'hit/'
                    stand = 'stand'
                    while True:
                        try: 
                            next = raw_input('{0}, your next move is..?:({1}{2})'.format(p.name,hit,stand)).lower()
                        except Exception,e:
                            print str(e)
                        else:
                            if next not in ['hit','stand']:
                                continue
                            else:
                                if next == 'hit':
                                    p.hit()
                                    break
                                elif next == 'stand':
                                    p.stand()
                                    break

def check_next():
    global desk_in_order
    #check dealer is busted
    dealer = desk_in_order[-1]
    if dealer.status == 'busted':
        return False
    else:
        #check is there any player alive    
        alive_count = 0
        for p in desk_in_order:
            if p.status == 'alive' and p.role == 'player':
                alive_count += 1
        return alive_count > 0
    
def check_chips():
    global desk_in_order
    for p in desk_in_order:
        if p.chips == 0 and p.role == 'player':
            desk_in_order.remove(p)
            print 'Player {0}, you are running out of chips.\nPlayer {0} been kicked out.'.format(p.name) 
            
    for p in desk_in_order:
        if p.chips == 0 and p.role == 'dealer':
            print 'Dealer was broke. Players please come later.'
            return True
    
    if(check_player_left(desk_in_order) == 0):
        return True
    else:
        return False

def check_player_left(desk_in_order):
    left = 0
    for p in desk_in_order:
        if p.role == 'player':
            left += 1
    return left
            
def reset():
    global desk_in_order,ful_deck
    for p in desk_in_order:
        p.card = []
        p.points = 0
        p.status = 'alive'
        ful_deck = getDecks()
        
        
    
```

```python


def game_on():   
    global desk_in_order, ful_deck,game_status
    
    try:
        while True:
            try:
                player_count = int(raw_input('How many players?(1-8):'))
            except:
                print 'Please advise integer only'
            else:
                if player_count > 8:
                    print 'Too many players.Please advise.'
                    continue
                else:
                    for i in range(0,player_count):
                        add_player(i)
                    break
        
        desk_in_order.append(Dealer('Bank'));
        
        counter = 1
        
        while True:
            
            try:                
                print '===========================ROUND:{0}==========================='.format(counter)
                #round 1
                ask_bit()
                
                for i in range(0,2):
                    for p in desk_in_order:
                        p.hit()
                game_status = 2

                #round 2
                while True:            

                    if(check_next()):
                        print_board(False)
                        ask_player_for_next()
                        game_status +=1
                    else:
                        break


                message = round_counter()
                print_board(True)
                print '\n'+ message

                if(check_chips()):
                    print 'Game over.'
                    break
                counter += 1

            except KeyboardInterrupt:
                print 'interrupted.'
                
            finally:
                reset()
                game_status = 1
                
    except KeyboardInterrupt:
        print 'interrupted.'
    
    finally:
        desk_in_order = []
                

        
        
```
