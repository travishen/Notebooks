```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips['%'] = tips['tip'] / tips['total_bill'] * 100
tips.head()
```

```python
sns.lmplot('size','%',tips,x_estimator=np.mean)
```

```python
sns.lmplot('total_bill','%',tips,hue='sex',markers=['o','x'])
```

```python
sns.lmplot('total_bill','%',tips,hue='day')
```
